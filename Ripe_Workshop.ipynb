# How To make a Sankey Graph

```python
df_meta_c = df.query("dst_name == 'www.meta.com'")
source_meta = []
for x in df_meta_c['prb_id'].unique():
    source_meta.append(str(x))

for x in df_meta_c['state'].unique():
    source_meta.append(x)

id_pb = [str(x) for x in df_meta_c['prb_id'].unique()]

source2 = []
target2 = []
value2 = []

for idp in id_pb:
    df_temp = df_meta_c.query(f"prb_id == {int(idp)}").groupby(['state']).size().reset_index(name='count')
    for idx,row in df_temp.iterrows():
        source2.append(source_meta.index(idp))
        target2.append(source_meta.index(row['state']))
        value2.append(row['count'])

```

```python
fig = go.Figure(data=[go.Sankey(
    node = dict(
      pad = 15,
      thickness = 20,
      line = dict(color = "black", width = 0.1),
      label = source_meta,
      color = "blue"
    ),
    link = dict(
      source = source2, # indices correspond to labels, eg A1, A2, A1, B1, ...
      target = target2,
      value = value2
  ))])

fig.update_layout(title_text="META DST STATE Sankey Diagram", font_size=10)
fig.show()
```

```python
probe_ids = list(df['prb_id'].unique())
import requests

api_key = "3521bab5-8c9a-445c-80ee-d96cbf8881db"
base_url = "https://atlas.ripe.net/api/v2"
# we need to get the iformantion from all the probes used
probes_info = []
for probe in probe_ids:
    # Construct the API URL
    url = f"https://atlas.ripe.net/api/v2/probes/{probe}/"
    # Set up the headers with your API key
    headers = {
        'Authorization': f'Key {api_key}'
    }
    # Make the request
    response = requests.get(url, headers=headers)
    if response.status_code == 200:
        #print(response.json())
        probes_info.append(response.json())
    else :
        print(f'not able to find probe : {probe} ')
```

```python
traceroutes_data = []

for idx,row in df_probe_temp.iterrows():
    temp = df_meta_map.query(f"prb_id == {row['id']}").groupby(['name','org_list','asn_list','country','state','city','lat','long']).size().reset_index(name='count')
    # lets iterate over the query data
    for idx2,row2 in temp.iterrows():
      traceroutes_data.append(
          {
              'probe_id': row['id'],
              'locations': [
                  {'coordinates': [row['lat'],row['lon']], 'asn': row['asn'], 'org': row['description'], 'city': 'UNKOWN'}, #start
                  {'coordinates': [row2['lat'],row2['long'] ], 'asn': row2['asn_list'], 'org':row2['org_list']+'-'+row2['country'], 'city':row2['city']},#End
              ],
              'color': row['color']  # Color for traceroute 1
          }
      )
```

```python
traceroutes_data[0]
```

```python
import folium

# Datos de ejemplo
# Calcular el centro del mapa
all_locations = [location for traceroute in traceroutes_data for location in traceroute['locations']]
center_lat = sum([loc["coordinates"][0] for loc in all_locations]) / len(all_locations)
center_lon = sum([loc["coordinates"][1] for loc in all_locations]) / len(all_locations)

# Inicializar el mapa
map = folium.Map(location=[center_lat, center_lon], zoom_start=6)

# Función para agregar marcadores con colores distintos para origen y destino
def add_traceroute_markers(locations):
    for i, location in enumerate(locations):
        tooltip_html = f"""
        <b>ASN:</b> {location['asn']}<br>
        <b>Org:</b> {location['org']}<br>
        <b>City:</b> {location['city']}<br>
        """
        # Determinar el color del marcador: azul para origen, rojo para destino
        color = "blue" if i == 0 else "red"
        folium.Marker(
            location["coordinates"],
            tooltip=tooltip_html,
            icon=folium.Icon(color=color)  # Asignar color al marcador
        ).add_to(map)

# Función para agregar una línea entre los puntos
def add_traceroute_polyline(locations, color):
    coordinates = [loc["coordinates"] for loc in locations]
    folium.PolyLine(coordinates, color=color, weight=2.5, opacity=1).add_to(map)

# Agregar marcadores y polilíneas para cada traceroute
for traceroute in traceroutes_data:
    add_traceroute_polyline(traceroute['locations'], traceroute['color'])
    add_traceroute_markers(traceroute['locations'])

# Agregar control de capas
folium.LayerControl().add_to(map)

# Mostrar el mapa
map
```

```python
import folium
import random
from folium import plugins

# Sample traceroutes data (replace with your actual data)
#traceroutes_data = []
# Function to generate a random color
def generate_random_color():
    color = "#{:06x}".format(random.randint(0, 0xFFFFFF))
    return color

# Initialize the map
map = folium.Map(location=[0, 0], zoom_start=2)

# Create a feature group for each probe
probe_layers = {}

# Function to add markers and polyline for each traceroute
for traceroute in traceroutes_data:
    probe_id = traceroute['probe_id']
    color = generate_random_color()

    # Create a feature group for the probe
    probe_layers[probe_id] = folium.FeatureGroup(name=f'Probe {probe_id}')
    map.add_child(probe_layers[probe_id])

    # Add markers and polyline to the feature group
    for location in traceroute['locations']:
        tooltip_html = f"""
        <b>ASN:</b> {location['asn']}<br>
        <b>Org:</b> {location['org']}<br>
        <b>City:</b> {location['city']}<br>
        <b>Hop:</b> {location['hop']}<br>
        """
           # Define CSS styles for the square icon
        css_style = f"""
        background-color: {color};
        color: white;
        width: 24px;
        line-height: 24px;
        text-align: center;
        border-radius: 4px;
        """
        folium.Marker(
            location["coordinates"],
            tooltip=tooltip_html,
            icon=folium.DivIcon(html=f"<div style='{css_style}'>{location['hop']}</div>")
        ).add_to(probe_layers[probe_id])

    coordinates = [loc["coordinates"] for loc in traceroute['locations']]
    folium.PolyLine(coordinates, color=color, weight=2.5, opacity=1).add_to(probe_layers[probe_id])

# Add layer control to toggle visibility of each probe
folium.LayerControl().add_to(map)

# Display the map
map.save('map_ixsy_with_routes2.html')
map
```
