```python
probes_selected = [6398,19709,51739,60909,1004112]
```

```python
#Metodo para crear red de nodos desde una medicion
```

```python
source = []
Destination = []

for probe_id in probes_selected:
    df_probe = df.query("prb_id == {}".format(probe_id))
    for id,row in df_probe.iterrows():
      source.append(row["prb_id"])
      Destination.append(row["dst_addr"])

df_result = pd.DataFrame({"inicio":source,"fin":Destination})

```

```python
df_result
```

```python
# Crear un grafo a partir de un Dataframe
# ======================================================================================
fig, ax = plt.subplots(figsize=(10,7))
G = nx.from_pandas_edgelist(df_result, source="inicio", target="fin")
nx.draw(G, with_labels=False, ax=ax)
```

```python
#Grafo Bipartito
df_result["fin"].unique().tolist()
```

```python
# Crear un grafo bipartito vacÃ­o
G_peliculas_actores = nx.Graph()

# Agregar los nodos de cada grupo
G_peliculas_actores.add_nodes_from(["Pelicula_1", "Pelicula_2", "Pelicula_3"], bipartite="Peliculas")
G_peliculas_actores.add_nodes_from(["Actor_1", "Actor_2", "Actor_3"], bipartite="Actores")

# Agregar las relaciones entre los nodos
G_peliculas_actores.add_edges_from(
    [
        ("Pelicula_1", "Actor_1"),
        ("Pelicula_2", "Actor_2"),
        ("Pelicula_3", "Actor_3"),
        ("Pelicula_2", "Actor_3"),
        ("Pelicula_2", "Actor_1"),
    ]
)
```

```python
G_peliculas_actores.nodes(data=True)
```

```python
tipo_nodo = [
    G_peliculas_actores.nodes[i]["bipartite"]
    for i in G_peliculas_actores.nodes()
]
tipo_nodo
```

```python
# colores para cada tipo de nodo en el grafo bipartito
colores = {"Peliculas": "red", "Actores": "blue"}
colores_nodos = [colores[n] for n in tipo_nodo]

fig, ax = plt.subplots(figsize=(5, 4))
nx.draw(
    G_peliculas_actores,
    pos=nx.spring_layout(G_peliculas_actores),
    with_labels=True,
    node_color=colores_nodos,
    ax=ax,
)
```

```python
# Crear un grafo bipartito vacÃ­o
G_peliculas_actores2 = nx.Graph()

# Agregar los nodos de cada grupo
G_peliculas_actores2.add_nodes_from(probes_selected, bipartite="Probes")
G_peliculas_actores2.add_nodes_from(df_result["fin"].unique().tolist(), bipartite="Dir_ip")
```

```python
# Crear un grafo bipartito vacÃ­o
#G_peliculas_actores2 = nx.from_pandas_edgelist(df_result, source="inicio", target="fin")
#G_peliculas_actores2.add_edges_from(nx.from_pandas_edgelist(df_result, source="inicio", target="fin"))
edge_list_nodes = []
for id,row in df_result.iterrows():
  edge_list_nodes.append((row["inicio"],row["fin"]))
```

```python
G_peliculas_actores2.add_edges_from(edge_list_nodes)
```

```python
G_peliculas_actores2.nodes(data=True)
```

```python
tipo_nodo = [
    G_peliculas_actores2.nodes[i]["bipartite"]
    for i in G_peliculas_actores2.nodes()
]
tipo_nodo
```

```python
# colores para cada tipo de nodo en el grafo bipartito
colores = {"Probes": "red", "Dir_ip": "blue"}
colores_nodos = [colores[n] for n in tipo_nodo]

fig, ax = plt.subplots(figsize=(10, 7))
nx.draw(
    G_peliculas_actores2,
    pos=nx.spring_layout(G_peliculas_actores2),
    with_labels=False,
    node_color=colores_nodos,
    ax=ax,
)
```
